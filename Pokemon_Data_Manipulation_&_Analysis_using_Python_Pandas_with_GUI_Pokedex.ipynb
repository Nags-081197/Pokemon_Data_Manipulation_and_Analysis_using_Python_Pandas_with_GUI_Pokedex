```python
import urllib.request
from PIL import Image, ImageTk
import tkinter as tk
import re
import pandas as pd
```

```python
# pokedex modelling

def pokemonCard(data,ind):
    url = 'https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/'+ str(data['#'][ind])+'.png'
    urllib.request.urlretrieve(url,"card_img.png")
    img = Image.open("card_img.png")
    
    # Tkinter
    root = tk.Tk()
    root.title("Pokedex")
    
    canvas = tk.Canvas(root, width = 350, height = 150)
    canvas.grid(columnspan = 3)
    
    #pokemon image 
    pokemon_img = Image.open("card_img.png")
    pokemon_img = ImageTk.PhotoImage(pokemon_img)
    pokemon_img_label = tk.Label(root,image = pokemon_img)
    pokemon_img_label.image = pokemon_img
    pokemon_img_label.grid(column = 1, row = 0)
    
    #pokemon information 
    Pokedex_Number_label = tk.Label(root, text =  "Pokedex No -")
    Pokedex_Number_label.grid(columnspan = 2, column = 0, row = 1)
    
    Pokedex_Number = tk.Label(root, text =str(data['#'][ind]))
    Pokedex_Number.grid(columnspan = 2, column = 1, row = 1)
    
    
    Pokedex_Name_out = tk.Label(root, text = "Name -")
    Pokedex_Name_out.grid(columnspan = 2, column = 0, row = 2)
    
    Pokedex_Name = tk.Label(root, text = str(data['Name'][ind]))
    Pokedex_Name.grid(columnspan = 2, column = 1, row = 2)
    
    
    Pokedex_type1_out = tk.Label(root, text = "Type 1 -")
    Pokedex_type1_out.grid(columnspan = 2, column = 0, row = 3)
    
    Pokedex_type1 = tk.Label(root, text = str(data['Type 1'][ind]))
    Pokedex_type1.grid(columnspan = 2, column = 1, row = 3)
    
    
    Pokedex_type2_out = tk.Label(root, text = "Type 2 -")
    Pokedex_type2_out.grid(columnspan = 2, column = 0, row = 4)
    
    Pokedex_type2 = tk.Label(root, text =str(data['Type 2'][ind]))
    Pokedex_type2.grid(columnspan = 2, column = 1, row = 4)
    
    
    Pokedex_hp_out = tk.Label(root, text = "HP -")
    Pokedex_hp_out.grid(columnspan = 2, column = 0, row = 5)
    
    Pokedex_hp = tk.Label(root, text =str(data['HP'][ind]))
    Pokedex_hp.grid(columnspan = 2, column = 1,row = 5)
    
    
    Pokedex_attack_out = tk.Label(root, text = "Attack -")
    Pokedex_attack_out.grid(columnspan = 2, column = 0, row = 6)
    
    Pokedex_attack = tk.Label(root, text = str(data['Attack'][ind]))
    Pokedex_attack.grid(columnspan = 2, column = 1, row = 6)
    
    
    Pokedex_defense_out = tk.Label(root, text = "Defense -")
    Pokedex_defense_out.grid(columnspan = 2, column = 0, row = 7)
    
    Pokedex_defense = tk.Label(root, text = str(data['Defense'][ind]))
    Pokedex_defense.grid(columnspan = 2, column = 1, row = 7)
    
    
    Pokedex_splatk_out = tk.Label(root, text = "Sp. Atk -")
    Pokedex_splatk_out.grid(columnspan = 2, column = 0, row = 8)
    
    Pokedex_splatk = tk.Label(root, text = str(data['Sp. Atk'][ind]))
    Pokedex_splatk.grid(columnspan = 2, column = 1, row = 8)
    
    
    Pokedex_spldef_out = tk.Label(root, text = "Sp. Def -")
    Pokedex_spldef_out.grid(columnspan = 2, column = 0, row = 9)
    
    Pokedex_spldef = tk.Label(root, text = str(data['Sp. Def'][ind]))
    Pokedex_spldef.grid(columnspan = 2, column = 1, row = 9)
    
    
    Pokedex_speed_out = tk.Label(root, text = "Speed -")
    Pokedex_speed_out.grid(columnspan = 2, column = 0, row = 10)
    
    Pokedex_speed = tk.Label(root, text = str(data['Speed'][ind]))
    Pokedex_speed.grid(columnspan = 2, column = 1, row = 10)
    
    
    Pokedex_Generation_out = tk.Label(root, text = "Generation -")
    Pokedex_Generation_out.grid(columnspan = 2, column = 0, row = 11)
    
    Pokedex_Generation = tk.Label(root, text = str(data['Generation'][ind]))
    Pokedex_Generation.grid(columnspan = 2, column = 1, row = 11)
    
    
    Pokedex_Legendary_out = tk.Label(root, text = "Legendary -")
    Pokedex_Legendary_out.grid(columnspan = 2, column = 0, row = 12)
    
    Pokedex_Legendary = tk.Label(root, text = str(data['Legendary'][ind]))
    Pokedex_Legendary.grid(columnspan = 2, column = 1, row = 12)
    
    Pokedex_space= tk.Label(root, text = " ")
    Pokedex_space.grid(columnspan = 3, column = 0, row = 14)

    root.mainloop()
    
    
    
    


dataframe = pd.read_csv('modified_pokemon_data.csv')

pokemon_name = input("Enter Pokemon Name : ").strip()

new_df = dataframe.loc[dataframe['Name'].str.contains(pokemon_name, flags = re.I, regex=True)]

if not new_df.empty:
    ind = dataframe.loc[dataframe['Name'].str.contains(pokemon_name, flags = re.I, regex=True)].index[0]
    pokemonCard(new_df,ind)
else:
    print('Pokemon not present in dataset')
    


    
                                                                                                                
```
